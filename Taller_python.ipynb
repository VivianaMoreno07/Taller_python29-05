```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos
X = np.linspace(-10, 10, 100)
Y = np.log(X + np.sin(np.pi / 2)) + np.sqrt(np.arctan((X / 5) + np.pi / 8))

# Dividir los datos en segmentos
num_segments = 5
segment_length = len(X) // num_segments

# Crear el gr√°fico con l√≠neas de diferentes colores
for i in range(num_segments):
    start_idx = i * segment_length
    end_idx = (i + 1) * segment_length if i < num_segments - 1 else len(X)
    plt.plot(X[start_idx:end_idx], Y[start_idx:end_idx])

# Agregar t√≠tulo y etiquetas de ejes
plt.title('Gr√°fico de la funci√≥n')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

# Activar la cuadr√≠cula
plt.grid(True)

# Mostrar el gr√°fico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generar datos
X = np.linspace(-10, 10, 100)
Y = np.log(X + np.sin(np.pi / 2)) + np.sqrt(np.arctan((X / 5) + np.pi / 8))
Z = np.sin(X) + np.cos(Y)

# Dividir los datos en tres segmentos
segment_length = len(X) // 3

# Crear figura y ejes 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Trazar cada segmento con un color diferente
ax.plot(X[:segment_length], Y[:segment_length], Z[:segment_length], color='blue')
ax.plot(X[segment_length:2*segment_length], Y[segment_length:2*segment_length], Z[segment_length:2*segment_length], color='green')
ax.plot(X[2*segment_length:], Y[2*segment_length:], Z[2*segment_length:], color='red')

# Configurar etiquetas de ejes
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')

# Mostrar el gr√°fico
plt.show()
```
